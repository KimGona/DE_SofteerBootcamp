```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import time

# ✅ 크롬 옵션 설정
options = webdriver.ChromeOptions()
#options.add_argument("--headless")  # 브라우저 없이 실행
#options.add_argument("--disable-gpu")
# options.add_argument("--no-sandbox")
# options.add_argument("--disable-dev-shm-usage")

# ✅ 최신 ChromeDriver 자동 설치
service = Service(ChromeDriverManager().install())

# # ✅ WebDriver 실행
# driver = webdriver.Chrome(service=service, options=options)

# WebDriver 초기화
driver = webdriver.Chrome(options=options)

# ✅ 크롤링할 URL
url = "https://1xbet.whoscored.com/players/148503/show/ollie-watkins"
driver.get(url)

# ✅ 버튼 클릭 
try:
    # 'sub-navigation' 선택
    sub_navigation = WebDriverWait(driver, 5).until(
        EC.element_to_be_clickable((By.XPATH, '//*[@id="sub-navigation"]/ul/li[1]/a'))
    )
    sub_navigation.click()
    time.sleep(2)  # 페이지 로딩 대기

    # 'player-tournament-stats-options' 선택
    player_stats_option = WebDriverWait(driver, 5).until(
        EC.element_to_be_clickable((By.XPATH, '//*[@id="player-tournament-stats-options"]/li[1]/a'))
    )
    player_stats_option.click()
    time.sleep(2)

    # 'field' 선택
    field_option = WebDriverWait(driver, 5).until(
        EC.element_to_be_clickable((By.XPATH, '//*[@id="field"]/dd[1]/a'))
    )
    field_option.click()
    time.sleep(2)

    # ✅ HTML 데이터 가져오기
    soup = BeautifulSoup(driver.page_source, "html.parser")
    #stats_div = soup.find("div", {"id": "statistics-table-summary"})
    stats_div = soup.find("div", {"class": "ml12-lg-3 ml12-m-4 ml12-s-5 ml12-xs-6 semi-attached-table"})
    # ✅ 결과 출력
    if stats_div:
        print(stats_div.prettify())
    else:
        print("❌ 해당 div를 찾을 수 없습니다.")

except Exception as e:
    print(f"❌ 오류 발생: {e}")

# # ✅ WebDriver 종료
# driver.quit()
```

```python
import pandas as pd

# 테이블 찾기
table = stats_div.find('table')
if table:
    rows = table.find_all('tr')  # 모든 행 가져오기

    # 데이터 저장용 리스트
    data = []

    # 헤더 추출
    headers = [th.get_text(strip=True) for th in rows[0].find_all('th')]

    # 데이터 행 추출
    for row in rows[1:]:
        cells = row.find_all('td')
        if len(cells) > 0:
            data.append([cell.get_text(strip=True) for cell in cells])

    # DataFrame 생성
    df = pd.DataFrame(data, columns=headers)
    print(df)
else:
    print("No table found.")

# ✅ WebDriver 종료
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import time

# ✅ 크롬 옵션 설정
options = webdriver.ChromeOptions()
#options.add_argument("--headless")  # 브라우저 없이 실행
#options.add_argument("--disable-gpu")
# options.add_argument("--no-sandbox")
# options.add_argument("--disable-dev-shm-usage")

# ✅ 최신 ChromeDriver 자동 설치
service = Service(ChromeDriverManager().install())

# # ✅ WebDriver 실행
# driver = webdriver.Chrome(service=service, options=options)

# WebDriver 초기화
driver = webdriver.Chrome(options=options)

# ✅ 크롤링할 URL
url = "https://1xbet.whoscored.com/players/148503/show/ollie-watkins"
driver.get(url)

# ✅ HTML 데이터 가져오기
soup = BeautifulSoup(driver.page_source, "html.parser")
#stats_div = soup.find("div", {"id": "statistics-table-summary"})
stats_div = soup.find("div", {"id": "player-matches-table"})
#✅ 결과 출력
# if stats_div:
#     print(stats_div.prettify())
# else:
#     print("❌ 해당 div를 찾을 수 없습니다.")


# import pandas as pd
# from bs4 import BeautifulSoup

# html = """[여기에 HTML 코드 삽입]"""  # HTML 코드 넣기

# # BeautifulSoup으로 HTML 파싱
# soup = BeautifulSoup(html, "html.parser")

# 데이터를 저장할 리스트
data = []

# 각 경기 데이터 가져오기
matches = stats_div.find_all("div", class_="divtable-row")
#print(matches)
for match in matches:
    try:
        # 날짜 가져오기
        date_tag = match.find("div", class_="col12-lg-1 col12-m-2 col12-s-0 col12-xs-0 divtable-data date-long")
        date = date_tag.text.strip() if date_tag else "N/A"

        # 홈팀 가져오기
        home_team_tag = match.find("div", class_="home-team")
        home_team = home_team_tag.text.strip() if home_team_tag else "N/A"

        # 원정팀 가져오기
        away_team_tag = match.find("div", class_="away-team")
        away_team = away_team_tag.text.strip() if away_team_tag else "N/A"

        # 경기 스코어 가져오기
        score_tag = match.find("a", class_="horiz-match-link")
        score = score_tag.text.strip() if score_tag else "N/A"

        # 출전 시간 가져오기
        minutes_tag = match.find("div", class_="col-data-mins")
        minutes = minutes_tag.text.strip() if minutes_tag else "N/A"

        # 평점 가져오기
        rating_tag = match.find("div", class_="col-data-rating")
        rating = rating_tag.text.strip() if rating_tag else "N/A"

        # 데이터 리스트에 추가
        data.append([date, home_team, away_team, score, minutes, rating])

    except Exception as e:
        print(f"❌ 오류 발생: {e}")

# Pandas DataFrame 생성
df = pd.DataFrame(data, columns=["Date", "Home Team", "Away Team", "Score", "Played Minutes", "Rating"])

# 결과 출력
print(df)

```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import time

# ✅ 크롬 옵션 설정
options = webdriver.ChromeOptions()
#options.add_argument("--headless")  # 브라우저 없이 실행
#options.add_argument("--disable-gpu")
# options.add_argument("--no-sandbox")
# options.add_argument("--disable-dev-shm-usage")

# ✅ 최신 ChromeDriver 자동 설치
service = Service(ChromeDriverManager().install())

# # ✅ WebDriver 실행
# driver = webdriver.Chrome(service=service, options=options)

# WebDriver 초기화
driver = webdriver.Chrome(options=options)

# ✅ 크롤링할 URL
url = "https://1xbet.whoscored.com/players/148503/show/ollie-watkins"
driver.get(url)

# ✅ HTML 데이터 가져오기
soup1 = BeautifulSoup(driver.page_source, "html.parser")
#stats_div = soup.find("div", {"id": "statistics-table-summary"})
soup = soup1.find("div", {"class": "sws-content character-card singular"})
#print(stats_div)

# BeautifulSoup으로 HTML 파싱
#soup = BeautifulSoup(html, "html.parser")
#print(soup)

import pandas as pd

# Strengths 데이터 추출
strengths = []
for strength in soup.select(".strengths .character"):
    attr = strength.select_one(".iconize-icon-left").text.strip()
    level = strength.select_one(".sws-level-wrapper").text.strip()
    strengths.append([attr, level])

# Weaknesses 데이터 추출
weaknesses = []
for weakness in soup.select(".weaknesses .character"):
    attr = weakness.select_one(".iconize-icon-left").text.strip()
    level = weakness.select_one(".sws-level-wrapper").text.strip()
    weaknesses.append([attr, level])

# Style of Play 데이터 추출
styles = []
for style in soup.select(".style .character"):
    styles.append([style.text.strip()])

# 데이터프레임 생성
df_strengths = pd.DataFrame(strengths, columns=["Attribute", "Level"])
df_weaknesses = pd.DataFrame(weaknesses, columns=["Attribute", "Level"])
df_styles = pd.DataFrame(styles, columns=["Style of Play"])

# 결과 출력
print("📌 Strengths")
print(df_strengths)
print("\n📌 Weaknesses")
print(df_weaknesses)
print("\n📌 Style of Play")
print(df_styles)
```
