```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import time

# ‚úÖ ÌÅ¨Î°¨ ÏòµÏÖò ÏÑ§Ï†ï
options = webdriver.ChromeOptions()
#options.add_argument("--headless")  # Î∏åÎùºÏö∞Ï†Ä ÏóÜÏù¥ Ïã§Ìñâ
#options.add_argument("--disable-gpu")
# options.add_argument("--no-sandbox")
# options.add_argument("--disable-dev-shm-usage")

# ‚úÖ ÏµúÏã† ChromeDriver ÏûêÎèô ÏÑ§Ïπò
service = Service(ChromeDriverManager().install())

# # ‚úÖ WebDriver Ïã§Ìñâ
# driver = webdriver.Chrome(service=service, options=options)

# WebDriver Ï¥àÍ∏∞Ìôî
driver = webdriver.Chrome(options=options)

# ‚úÖ ÌÅ¨Î°§ÎßÅÌï† URL
url = "https://1xbet.whoscored.com/players/148503/show/ollie-watkins"
driver.get(url)

# ‚úÖ HTML Îç∞Ïù¥ÌÑ∞ Í∞ÄÏ†∏Ïò§Í∏∞
soup = BeautifulSoup(driver.page_source, "html.parser")
#stats_div = soup.find("div", {"id": "statistics-table-summary"})
stats_div = soup.find("div", {"id": "player-matches-table"})
#‚úÖ Í≤∞Í≥º Ï∂úÎ†•
# if stats_div:
#     print(stats_div.prettify())
# else:
#     print("‚ùå Ìï¥Îãπ divÎ•º Ï∞æÏùÑ Ïàò ÏóÜÏäµÎãàÎã§.")


# import pandas as pd
# from bs4 import BeautifulSoup

# html = """[Ïó¨Í∏∞Ïóê HTML ÏΩîÎìú ÏÇΩÏûÖ]"""  # HTML ÏΩîÎìú ÎÑ£Í∏∞

# # BeautifulSoupÏúºÎ°ú HTML ÌååÏã±
# soup = BeautifulSoup(html, "html.parser")

# Îç∞Ïù¥ÌÑ∞Î•º Ï†ÄÏû•Ìï† Î¶¨Ïä§Ìä∏
data = []

# Í∞Å Í≤ΩÍ∏∞ Îç∞Ïù¥ÌÑ∞ Í∞ÄÏ†∏Ïò§Í∏∞
matches = stats_div.find_all("div", class_="divtable-row")
#print(matches)
for match in matches:
    try:
        # ÎÇ†Ïßú Í∞ÄÏ†∏Ïò§Í∏∞
        date_tag = match.find("div", class_="col12-lg-1 col12-m-2 col12-s-0 col12-xs-0 divtable-data date-long")
        date = date_tag.text.strip() if date_tag else "N/A"

        # ÌôàÌåÄ Í∞ÄÏ†∏Ïò§Í∏∞
        home_team_tag = match.find("div", class_="home-team")
        home_team = home_team_tag.text.strip() if home_team_tag else "N/A"

        # ÏõêÏ†ïÌåÄ Í∞ÄÏ†∏Ïò§Í∏∞
        away_team_tag = match.find("div", class_="away-team")
        away_team = away_team_tag.text.strip() if away_team_tag else "N/A"

        # Í≤ΩÍ∏∞ Ïä§ÏΩîÏñ¥ Í∞ÄÏ†∏Ïò§Í∏∞
        score_tag = match.find("a", class_="horiz-match-link")
        score = score_tag.text.strip() if score_tag else "N/A"

        # Ï∂úÏ†Ñ ÏãúÍ∞Ñ Í∞ÄÏ†∏Ïò§Í∏∞
        minutes_tag = match.find("div", class_="col-data-mins")
        minutes = minutes_tag.text.strip() if minutes_tag else "N/A"

        # ÌèâÏ†ê Í∞ÄÏ†∏Ïò§Í∏∞
        rating_tag = match.find("div", class_="col-data-rating")
        rating = rating_tag.text.strip() if rating_tag else "N/A"

        # Îç∞Ïù¥ÌÑ∞ Î¶¨Ïä§Ìä∏Ïóê Ï∂îÍ∞Ä
        data.append([date, home_team, away_team, score, minutes, rating])

    except Exception as e:
        print(f"‚ùå Ïò§Î•ò Î∞úÏÉù: {e}")

# Pandas DataFrame ÏÉùÏÑ±
df = pd.DataFrame(data, columns=["Date", "Home Team", "Away Team", "Score", "Played Minutes", "Rating"])

# Í≤∞Í≥º Ï∂úÎ†•
print(df)

```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import time

# ‚úÖ ÌÅ¨Î°¨ ÏòµÏÖò ÏÑ§Ï†ï
options = webdriver.ChromeOptions()
#options.add_argument("--headless")  # Î∏åÎùºÏö∞Ï†Ä ÏóÜÏù¥ Ïã§Ìñâ
#options.add_argument("--disable-gpu")
# options.add_argument("--no-sandbox")
# options.add_argument("--disable-dev-shm-usage")

# ‚úÖ ÏµúÏã† ChromeDriver ÏûêÎèô ÏÑ§Ïπò
service = Service(ChromeDriverManager().install())

# # ‚úÖ WebDriver Ïã§Ìñâ
# driver = webdriver.Chrome(service=service, options=options)

# WebDriver Ï¥àÍ∏∞Ìôî
driver = webdriver.Chrome(options=options)

# ‚úÖ ÌÅ¨Î°§ÎßÅÌï† URL
url = "https://1xbet.whoscored.com/players/148503/show/ollie-watkins"
driver.get(url)

# ‚úÖ HTML Îç∞Ïù¥ÌÑ∞ Í∞ÄÏ†∏Ïò§Í∏∞
soup1 = BeautifulSoup(driver.page_source, "html.parser")
#stats_div = soup.find("div", {"id": "statistics-table-summary"})
soup = soup1.find("div", {"class": "sws-content character-card singular"})
#print(stats_div)

# BeautifulSoupÏúºÎ°ú HTML ÌååÏã±
#soup = BeautifulSoup(html, "html.parser")
#print(soup)

import pandas as pd

# Strengths Îç∞Ïù¥ÌÑ∞ Ï∂îÏ∂ú
strengths = []
for strength in soup.select(".strengths .character"):
    attr = strength.select_one(".iconize-icon-left").text.strip()
    level = strength.select_one(".sws-level-wrapper").text.strip()
    strengths.append([attr, level])

# Weaknesses Îç∞Ïù¥ÌÑ∞ Ï∂îÏ∂ú
weaknesses = []
for weakness in soup.select(".weaknesses .character"):
    attr = weakness.select_one(".iconize-icon-left").text.strip()
    level = weakness.select_one(".sws-level-wrapper").text.strip()
    weaknesses.append([attr, level])

# Style of Play Îç∞Ïù¥ÌÑ∞ Ï∂îÏ∂ú
styles = []
for style in soup.select(".style .character"):
    styles.append([style.text.strip()])

# Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÏÉùÏÑ±
df_strengths = pd.DataFrame(strengths, columns=["Attribute", "Level"])
df_weaknesses = pd.DataFrame(weaknesses, columns=["Attribute", "Level"])
df_styles = pd.DataFrame(styles, columns=["Style of Play"])

# Í≤∞Í≥º Ï∂úÎ†•
print("üìå Strengths")
print(df_strengths)
print("\nüìå Weaknesses")
print(df_weaknesses)
print("\nüìå Style of Play")
print(df_styles)
```
